### Импорт библиотек

```python
import math
import time
from random import seed
from random import randint
```

```python
seed(1)
m = input('Enter columns ')
n = input('Enter rows ')
min_limit = input('Enter min value ')
max_limit = input('Enter max value ')

m = 50 if m == "" else int(m)
n = 50 if n == "" else int(n)
min_limit = -250 if min_limit == "" or int(max_limit) < int(min_limit) else int(min_limit)
max_limit = 1013 if max_limit == "" or int(max_limit) < int(min_limit) else int(max_limit)

# Функция генерации матрицы
def generateMatrix():
    matrix = [[0] * m for i in range(n)]

    for i in range(n):
        for j in range(m):
            matrix[i][j] = randint(min_limit, max_limit)

    return matrix
```

### Задание №3

### Перечисление функций:

```python
# Сортировка выбором
def selectionSort(matrix):

    for row in matrix:
        for i in range(len(row)):
            # Принимаем первый элемент минимальным
            min = i
            for j in range(i + 1, len(row)):
                if row[j] < row[min]:
                    min = j
            row[i], row[min] = row[min], row[i]

# Сортировка Вставкой
def insertionSort(matrix):

    for row in matrix:
        for i in range(1, len(row)):
            item_to_insert = row[i]
            j = i - 1

            while j >= 0 and row[j] > item_to_insert:
                row[j + 1] = row[j]
                j -= 1

            row[j + 1] = item_to_insert

# Сортировка Обменом (Пузырьковый метод)
def bubbleSort(matrix):
    for row in matrix:
        swapped = True
        while swapped:
            swapped = False
            for i in range(len(row) - 1):
                if row[i] > row[i + 1]:
                    row[i], row[i + 1] = row[i + 1], row[i]
                    swapped = True

# Сортировка Шелла
def shellSort(matrix):
    for row in matrix:
        n = len(row)
        k = int(math.log2(n))
        interval = 2**k -1
        while interval > 0:
            for i in range(interval, n):
                temp = row[i]
                j = i
                while j >= interval and row[j - interval] > temp:
                    row[j] = row[j - interval]
                    j -= interval
                row[j] = temp
            k -= 1
            interval = 2**k -1
    return matrix

# Турнирная сортировка

def tournamentSort(matrix):
    tree = [None] * 2 * (len(matrix) + len(matrix) % 2)
    index = len(tree) - len(matrix) - len(matrix) % 2

    for i, v in enumerate(matrix):
        tree[index + i] = (i, v)

    for j in range(len(matrix)):
        n = len(matrix)
        index = len(tree) - len(matrix) - len(matrix) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1)
                if tree[i] != None and tree[i + 1] != None:
                    if tree[i][1] < tree[i + 1][1]:
                        tree[i // 2] = tree[i]
                    else:
                        tree[i // 2] = tree[i + 1]
                else:
                    tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
            index -= n

        index, x = tree[0]
        matrix[j] = x
        tree[len(tree) - len(matrix) - len(matrix) % 2 + index] = None
def TournamentSort(matrix):
    for i in range(len(matrix)):
        tournamentSort(matrix[i])
    return matrix

# Быстрая сортировка
def partition(nums, low, high):
    pivot = nums[(low + high) // 2]
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while nums[i] < pivot:
            i += 1

        j -= 1
        while nums[j] > pivot:
            j -= 1

        if i >= j:
            return j

        nums[i], nums[j] = nums[j], nums[i]

def quickSort(matrix):
    def _quickSort(items, low, high):
        if low < high:
            split_index = partition(items, low, high)
            _quickSort(items, low, split_index)
            _quickSort(items, split_index + 1, high)
    for row in matrix:
        _quickSort(row, 0, len(row) - 1)



# Пирамидальная сортировка
def heapify(nums, heap_size, root_index):
    # Индекс наибольшего элемента считаем корневым индексом
    largest = root_index
    left_child = (2 * root_index) + 1
    right_child = (2 * root_index) + 2

    # Если левый потомок корня — допустимый индекс, а элемент больше,
    # чем текущий наибольший, обновляем наибольший элемент
    if left_child < heap_size and nums[left_child] > nums[largest]:
        largest = left_child

    # То же самое для правого потомка корня
    if right_child < heap_size and nums[right_child] > nums[largest]:
        largest = right_child

    # Если наибольший элемент больше не корневой, они меняются местами
    if largest != root_index:
        nums[root_index], nums[largest] = nums[largest], nums[root_index]
        # Heapify the new root element to ensure it's the largest
        heapify(nums, heap_size, largest)

def heapSort(matrix):
    
    for row in matrix:
        n = len(row)
        for i in range(n, -1, -1):
            heapify(row, n, i)

        for i in range(n - 1, 0, -1):
            row[i], row[0] = row[0], row[i]
            heapify(row, i, 0)

# Встроенная в язык сортировка
def standartSort(matrix):
    for row in matrix:
        row.sort()
```

### Результат и время выполнения всех алгоритмов

```python
print('1. Сортировка выбором')
matrix1 = generateMatrix()
TimeStart = time.time()
selectionSort(matrix1)
TimeEnd = time.time()
Delta = TimeEnd - TimeStart
print(Delta, '\n')

print('2. Сортировка вставкой')
matrix2 = generateMatrix()
TimeStart = time.time()
insertionSort(matrix2)
TimeEnd = time.time()
Delta = TimeEnd - TimeStart
print(Delta, '\n')

print('3. Сортировка Обменом (Пузырьковый метод)')
matrix3 = generateMatrix()
TimeStart = time.time()
bubbleSort(matrix3)
TimeEnd = time.time()
Delta = TimeEnd - TimeStart
print(Delta, '\n')

print('4. Сортировка Шелла')
matrix4 = generateMatrix()
TimeStart = time.time()
shellSort(matrix4)
TimeEnd = time.time()
Delta = TimeEnd - TimeStart
print(Delta, '\n')

print('5. Турнирная сортировка')
matrix5 = generateMatrix()
TimeStart = time.time()
TournamentSort(matrix5)
TimeEnd = time.time()
Delta = TimeEnd - TimeStart
print(Delta, '\n')

print('6. Быстрая сортировка')
matrix6 = generateMatrix()
TimeStart = time.time()
quickSort(matrix6)
TimeEnd = time.time()
Delta = TimeEnd - TimeStart
print(Delta, '\n')

print('7. Пирамидальная сортировка')
matrix7 = generateMatrix()
TimeStart = time.time()
heapSort(matrix7)
TimeEnd = time.time()
Delta = TimeEnd - TimeStart
print(Delta, '\n')

print('8. Встроенная в язык сортировка')
matrix8 = generateMatrix()
TimeStart = time.time()
standartSort(matrix8)
TimeEnd = time.time()
Delta = TimeEnd - TimeStart
print(Delta, '\n')
```
